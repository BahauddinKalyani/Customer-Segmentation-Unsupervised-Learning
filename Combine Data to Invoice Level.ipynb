```python
grouped = df.groupby('InvoiceNo')
```

```python
def get_not_null_value(df, column_name):
    not_null_values = df[df[column_name].notnull()][column_name]
    if not not_null_values.empty:
        return not_null_values.iloc[0]
    else:
        return np.nan

def process_grp(df):
    df['TotalPrice'] = df['UnitPrice'] * df['Quantity']
    Subtotal = df['TotalPrice'].sum()
    InvoiceNo = df['InvoiceNo'].iloc[0]
    InvoiceDate = df['InvoiceDate'].iloc[0]
    Country = df['Country'].iloc[0]
    CustomerID = get_not_null_value(df, 'CustomerID')

    return [InvoiceNo, InvoiceDate, CustomerID, Subtotal, Country]
    
```

```python
n=0
new_df = pd.DataFrame(columns=['InvoiceNo', 'InvoiceDate', 'CustomerID', 'Subtotal', 'Country'])
for group_name, group_df in grouped:
    if n==25901:
        break
    data = process_grp(group_df)
    new_df.loc[n] = data
    n+=1

new_df.head(10)
```
